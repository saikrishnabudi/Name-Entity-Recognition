# **Emotion Mining**

```python
# Sentimental Analysis - for this i am using the the Affin dictonary where it was telling the Negative Values telling the negative emotions and positive value telling the positive emotions
affin=pd.read_csv("/content/Afinn.csv",sep=",",encoding="latin-1")   # Encoding="latin1" helps to fix the special special symbols issues in the data
affin
```

```python
affin[1500:]
```

```python
import nltk
nltk.download("punkt")      # Punkt package is allows us to perform the Tokenization on the data
```

```python
#Performing the Sentence Tokenization
from nltk import tokenize
sentences=tokenize.sent_tokenize(" ".join(book))
sentences[10:15]
```

```python
# Seperating the sentences and form it into a Data Frame
sent_df=pd.DataFrame(sentences,columns=["sentence"])
sent_df
```

```python
# Converting the above DataFrame data into Dictonary.
# For every word has it's corresponding value based on the emotion word
affinity_scores=affin.set_index("word")["value"].to_dict()
```

```python
affinity_scores
```

```python
nlp=spacy.load("en_core_web_sm")
sentiment_lexicon=affinity_scores

def calculate_sentiment(text: str=None):
  sent_score=0
  if text:
    sentence=nlp(text)
    print(sentence)
    for word in sentence:
      sent_score+=sentiment_lexicon.get(word.lemma_,0)
  return sent_score
```

```python
calculate_sentiment(text="The child laughed")
```

```python
sent_df['sentiment_value']=sent_df['sentence'].apply(calculate_sentiment)
```
